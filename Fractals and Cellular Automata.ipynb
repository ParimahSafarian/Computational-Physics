### Q2 Cellular Automata 256 Rules

```python
rulenum = 110
n = 201
t = 200
line = np.zeros((t, n))
line[0, 101] = 1
line
```

```python
# How to plot any number's rule

def dec_to_bin(x):
    return int(bin(x)[2:])


def rules(rulenum, n, t):

    r = dec_to_bin(rulenum)
    line = np.zeros((t, n))
    line[0, int(n/2)+1] = 1
    
    for i in range(t-1):
        for j in range(n):
            if line[i, (j-1+n)%n] == 1:
                if line[i, j] == 1:
                    if line[i, (j+1)%n] == 1:
                        line[i+1, j] = int(r%1e8/1e7)
                    else:
                        line[i+1, j] = int(r%1e7/1e6)
                else:
                    if line[i, (j+1)%n] == 1:
                        line[i+1, j] = int(r%1e6/1e5)
                    else:
                        line[i+1, j] = int(r%1e5/1e4)
            else:
                if line[i, j] == 1:
                    if line[i, (j+1)%n] == 1:
                        line[i+1, j] = int(r%1e4/1e3)
                    else:
                        line[i+1, j] = int(r%1e3/1e2)
                else:
                    if line[i, (j+1)%n] == 1:
                        line[i+1, j] = int(r%1e2/1e1)
                    else:
                        line[i+1, j] = int(r%1e1/1e0)
    return line
```

```python
plt.imshow(rules(73, 201, 200), cmap = plt.cm.winter)
```

```python
#plot

fig, axs = plt.subplots(2, 2, figsize=(15,15))
axs[0, 0].imshow(rules(110, 201, 200))
axs[0, 0].set_title('110 Rule')
axs[0, 1].imshow(rules(75, 201, 200))
axs[0, 1].set_title('75 Rule')
axs[1, 0].imshow(rules(73, 201, 200))
axs[1, 0].set_title('73 Rule')
axs[1, 1].imshow(rules(90, 201, 200))
axs[1, 1].set_title('90 Rule')

for ax in axs.flat:
    ax.set(xlabel='n', ylabel='t')

# Hide x labels and tick labels for top plots and y ticks for right plots.
for ax in axs.flat:
    ax.label_outer()
    
plt.savefig("Q2-01.png")
```

### Q4 Game of Life

```python
t = 10

dynamics = np.zeros((6,6,t))
life2 = np.zeros((6,6))
life3 = np.zeros((6,6))
life4 = np.zeros((10,10))
```

```python
def alight_neighbours(life, i, j):
    alight = 0
    alight = np.sum(life[i-1, j], life[i+1, j], life[i, j-1], life[i, j+1], life[i-1, j-1],
                     life[i-1, j+1], life[i+1, j+1], life[i+1, j-1])
    return alight
```

```python
#initial conditions
life2[1, 2] = life2[1, 3] = life2[2, 1] = life2[2, 4] = life2[3, 2] = life2[3, 4] = life2[4, 3] = 1

plt.imshow(life2)
```

```python
for i in range(6):
    for j in range(6):
        alighted = alight_neighbours(life2, i, j)
        if life2[i, j] == 0:
            if alighted == 3:
                life2[i, j] == 1
        else:
            if alighted == 2 or alighted == 3:
                continue
            else:
                life2[i, j] = 0
```

```python
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np
import random

#creating graph
Life1 = nx.Graph()
Life1.add_nodes_from(range(36))

#add_edge
for j in range(36):
    Life1.add_edge(j, (j + 6)%6)
    Life1.add_edge(j, (j + 1)%6)

nx.set_node_attributes(Life1, values = 0, name = 'state')
nx.set_node_attributes(Life1, values = 0, name = 'future')

Life1.nodes[8]['state'] = Life1.nodes[9]['state'] = Life1.nodes[13]['state'] = Life1.nodes[16]['state'] = 1
Life1.nodes[20]['state'] = Life1.nodes[22]['state'] = Life1.nodes[27]['state'] = 1
```

```python
def Life(net, t, dynamics):
    for j in range(t):
        for i in range(np.shape(net)[0]):
            neighbors = list(net.neighbors(i))
            x = 0
            for k in neighbors:
                if net.nodes[k]['state'] == 1:
                    x = x + 1
            if net.nodes[i]['state'] == 1:
                if x == 2 or x == 3:
                    net.nodes[i]['future'] = 1
                else:
                    net.nodes[i]['future'] = 0
            else:
                if x == 3:
                    net.nodes[i]['future'] = 1
        for l in range(36):
            net.nodes[l]['state'] = net.nodes[l]['future']
        n = 0
        for i in range(int(np.sqrt(np.shape(net)[0]))):
            for m in range(int(np.sqrt(np.shape(net)[0]))):
                dynamics[i, m, j] = net.nodes[n]['state']
                #print(n)
                n = n + 1
    
    return dynamics
```
