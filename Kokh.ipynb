```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
#defining some function for scaling, rotaion and transition of vectors

#only scaling
def kokh1(in_x , in_y):
    x1 = np.array(1/3) * in_x
    y1 = np.array(1/3) * in_y
    return x1 , y1

#scaling ,rotation about 60 clockwise , translation
def kokh2(in_x , in_y):
    x2 = 1/3 * (in_x * np.array(np.cos(-np.pi/3)) + in_y * np.array(-np.sin(np.pi/3)) ) + 1/3 
    y2 = 1/3 * (in_y * np.array(np.cos(-np.pi/3)) - in_x * np.array(-np.sin(np.pi/3)) )
    return x2 , y2

#scaling, rotation about 60 anticlockwise, translation
def kokh3(in_x , in_y):
    x3 = 1/3 * ((in_x - np.array(1)) * np.array(np.cos(np.pi/3)) + in_y * np.array(np.sin(np.pi/3)) ) + 2/3 
    y3 = 1/3 * ( in_y * np.array(np.cos(np.pi/3)) - (in_x - np.array(1)) * np.array(np.sin(np.pi/3)) )
    return x3 , y3

#scaling, translation
def kokh4(in_x , in_y):
    x4 = np.array(1/3) * in_x + np.array(2/3)
    y4 = np.array(1/3) * in_y
    return x4 , y4
```

```python
#This function makes a kokh barfdane's 

def kokh(run_num, in_x, in_y):

    for i in range(run_num):
        x1 , y1 = kokh1(in_x , in_y )
        x2 , y2 = kokh2(in_x , in_y)
        x3 , y3 = kokh3(in_x , in_y)
        x4 , y4 = kokh4(in_x , in_y)

        in_x1 = np.append(x1 , x2) 
        in_x2 = np.append(x3 , x4)
        in_x = np.append(in_x1 , in_x2)
        in_y1 = np.append(y1 , y2)
        in_y2 = np.append(y3 , y4)
        in_y = np.append(in_y1 , in_y2)
        
    return in_x, in_y
```

```python
#Now just run...
run_num = 10

#initial values...
in_x = [0,1]
in_y = [0,0]

in_x, in_y = kokh(run_num, in_x , in_y)
#plot
fig, ax = plt.subplots()
ax.plot(in_x, in_y)
ax.set( title='Kokh' )
#to make the scale of axes equal
ax.axis('equal')
fig.savefig("test.png")
plt.show()
```
