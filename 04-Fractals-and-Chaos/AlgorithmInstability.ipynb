```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import random
import math
```

```python
#Constant values in SI units
v = 10
c = 1e-6
r = 3000
t_max = 0.04
```

```python
#number of steps 
step  = 100
h     = t_max / step
t = np.arange(0, t_max+h, h)

q = []
#initial value of q = 0
q.append(0)
q.append(q[0] + h * (v - q[0]/c)/r)
for n in range(1, step, 1):
    q.append(q[n-1] + 2 *  h * (v - q[n]/c)/r)
    
plt.plot(t, q, c = 'deepskyblue', alpha = 0.8)
plt.ylabel("$Q(t)$")
plt.xlabel("$t$")
plt.title(f"Charge\n h = {h}, using second order algorithm")

plt.savefig("charge2.png")
```

```python
#number of steps 
t_max = 0.02
step  = 100
h     = t_max / step
t = np.arange(0, t_max+h, h)

q = []
#initial value of q = 0
q.append(v*c)
q.append(q[0] + h * (v - q[0]/c)/r)
for n in range(1, step, 1):
    q.append(q[n-1] + 2 *  h * ( - q[n]/c)/r)
    
plt.plot(t, q, c = 'orange', alpha = 0.8)
plt.ylabel("$Q(t)$")
plt.xlabel("$t$")
plt.title(f"Decharge\n h = {h}, using second order algorithm")

plt.savefig("decharge2.png")
```
