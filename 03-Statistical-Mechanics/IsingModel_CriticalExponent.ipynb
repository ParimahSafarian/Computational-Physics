```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
import random
import pandas as pd
```

```python
corr_len100 = pd.read_csv('Data\correlation_m100.csv')[2:21]
corr_len130 = pd.read_csv('Data\correlation_m130.csv')[2:21]
corr_len150 = pd.read_csv('Data\correlation_m150.csv')[2:21]
corr_len180 = pd.read_csv('Data\correlation_m180.csv')[2:21]
corr_len200 = pd.read_csv('Data\correlation_m100.csv')[2:21]

energy_m100 = pd.read_csv('Data\energy_m100.csv')[2:21]
energy_m130 = pd.read_csv('Data\energy_m130.csv')[2:21]
energy_m150 = pd.read_csv('Data\energy_m150.csv')[2:21]
energy_m180 = pd.read_csv('Data\energy_m180.csv')[2:21]
energy_m200 = pd.read_csv('Data\energy_m200.csv')[2:21]

energy_var100 = pd.read_csv('Data\energy_var100.csv')[2:21]
energy_var130 = pd.read_csv('Data\energy_var130.csv')[2:21]
energy_var150 = pd.read_csv('Data\energy_var150.csv')[2:21]
energy_var180 = pd.read_csv('Data\energy_var180.csv')[2:21]
energy_var200 = pd.read_csv('Data\energy_var200.csv')[2:21]

magnetization_m100 = pd.read_csv('Data\magnetization_m100.csv')[2:21]
magnetization_m130 = pd.read_csv('Data\magnetization_m130.csv')[2:21]
magnetization_m150 = pd.read_csv('Data\magnetization_m150.csv')[2:21]
magnetization_m180 = pd.read_csv('Data\magnetization_m180.csv')[2:21]
magnetization_m200 = pd.read_csv('Data\magnetization_m200.csv')[2:21]

magnetization_var100 = pd.read_csv('Data\magnetization_var100.csv')[2:21]
magnetization_var130 = pd.read_csv('Data\magnetization_var130.csv')[2:21]
magnetization_var150 = pd.read_csv('Data\magnetization_var150.csv')[2:21]
magnetization_var180 = pd.read_csv('Data\magnetization_var180.csv')[2:21]
magnetization_var200 = pd.read_csv('Data\magnetization_var200.csv')[2:21]
```

```python
#beta values
beta_mat = np.append(np.append(np.arange(0.1, 0.4, 0.05), np.arange(0.41, 0.5, 0.01)), np.arange(0.505, 0.805, 0.05))[3:22]
beta_mat
```

```python
#plots
fig, axs = plt.subplots(2, 2, figsize=(17,17))
plt.suptitle("For l = 100", fontsize = 24)

#energy
axs[0, 0].errorbar(beta_mat, corr_len100, c = 'palevioletred')
axs[0, 0].set(xlabel='\u03B2', ylabel='\u03BE(\u03B2)')
axs[0, 0].set_title('$Correlation Lenght$')

#heat capacity

c_v = np.zeros(np.size(beta_mat))
for i in range(np.size(beta_mat)):
    c_v[i] = beta_mat[i] ** 2 * np.array(energy_var100)[i][0]
    
axs[0, 1].plot(beta_mat, c_v, c = 'deepskyblue')
axs[0, 1].set(xlabel='\u03B2', ylabel='$C_v(\u03B2)$')
axs[0, 1].set_title('$Heat Capacity$')

#mean magnetization
axs[1, 0].errorbar(beta_mat, magnetization_m100, c = 'sandybrown')
axs[1, 0].set(xlabel='\u03B2', ylabel='M(\u03B2)')
axs[1, 0].set_title('$MeanMagnetization$')

#susceptibility

chi = np.zeros(np.size(beta_mat))
for i in range(np.size(beta_mat)):
    chi[i] = beta_mat[i] * np.array(energy_var100)[i][0]
    
axs[1, 1].plot(beta_mat, chi, c = 'greenyellow')
axs[1, 1].set(xlabel='\u03B2', ylabel='\u03C7(\u03B2)')
axs[1, 1].set_title('$Susceptibility$')

plt.savefig("parameters100.png")
```

```python
corrlen_b = [0.49, 0.5, 0.505, 0.46]
cv_b      = [0.48, 0.45, 0.45, 0.44]
magnet_b  = [0.42, 0.35, 0.3, 0.4]
chi_b     = [0.48, 0.46, 0.45, 0.43]
l         = [100, 130, 150, 180]
```

Correlation lenght critical exponent

```python
fit1 = np.polyfit(np.log(l), np.log(corrlen_b), 1)

plt.plot(np.log(l), np.log(corrlen_b), '.', c = 'indigo', label = 'data')
plt.plot(np.log(l), fit1[0] * np.log(l) + fit1[1], c = 'gold', label = f'fit:exponent = {np.round(fit1[0], 3)}')
plt.title("Correlation lenght critical exponent")
plt.xlabel('$log(l)$')
plt.ylabel('$log(\u03B2_c)')
plt.legend()
plt.savefig('bood1.png')
```

```python
fit2 = np.polyfit(np.log(l), np.log(cv_b), 1)

plt.plot(np.log(l), np.log(cv_b), '.', c = 'indigo', label = 'data')
plt.plot(np.log(l), fit2[0] * np.log(l) + fit2[1], c = 'deepskyblue', label = f'fit:exponent = {np.round(fit2[0], 3)}')
plt.title("Heat capacity critical exponent")
plt.xlabel('$log(\u03B2_c)$')
plt.ylabel('$log(l)')
plt.legend()
plt.savefig('bood2.png')
```

```python
fit3 = np.polyfit(np.log(l), np.log(magnet_b), 1)

plt.plot(np.log(l), np.log(magnet_b), '.', c = 'lawngreen', label = 'data')
plt.plot(np.log(l), fit3[0] * np.log(l) + fit3[1], c = 'forestgreen', label = f'fit:exponent = {np.round(fit3[0], 3)}')
plt.title("Magnetization critical exponent")
plt.xlabel('$log(\u03B2_c)$')
plt.ylabel('$log(l)')
plt.legend()
plt.savefig('bood3.png')
```

```python
fit4 = np.polyfit(np.log(l), np.log(chi_b), 1)

plt.plot(np.log(l), np.log(chi_b), '.', c = 'black', label = 'data')
plt.plot(np.log(l), fit4[0] * np.log(l) + fit4[1], c = 'orangered', label = f'fit:exponent = {np.round(fit1[0], 3)}')
plt.title("Suusceptibility critical exponent")
plt.xlabel('$log(\u03B2_c)$')
plt.ylabel('$log(l)')
plt.legend()
plt.savefig('bood4.png')
```
