# Correlation Analysis

Statistical correlation analysis in Monte Carlo simulations. Studies temporal and spatial correlations in stochastic processes and their impact on sampling efficiency.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import random
import math
```

```python
 def n_after4(n):
    rand = []
    
    while np.size(rand) <= n:
        new_num = np.floor(random.random()*10)
        
        if new_num == 4:
            rand.append(np.floor(random.random()*10))
 
    return rand
```

```python
n = 2000
plt.hist(n_after4(n), 10, density=True, facecolor='lightpink')

plt.title("{} after 4 random number generated in range 0-9 histogram".format(n))
plt.ylabel("frequecy")
plt.xlabel("number")

plt.savefig("after4.png")
```
