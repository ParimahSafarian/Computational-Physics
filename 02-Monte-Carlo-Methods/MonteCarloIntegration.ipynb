```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import random
import math
import timeit
import pandas as pd
```

### Simple Sampling

```python
real_integral = 0.882081
a = 0
b = 2
points_mat = np.arange(1e5, 1e6+1e5, 1e5)
data = pd.DataFrame()
data = pd.DataFrame(columns = ['Number of points', 'Integral Value', 'Real Error', 'Statistical Error', 'Runtime'])

for points in points_mat:

    start    = timeit.default_timer()
    rand     = np.random.uniform(a, b, int(points))
    integral = np.mean(np.e ** (-rand ** 2)) * (b-a)
    error    = np.abs(integral - real_integral)
    stat_error = np.std(np.e ** (-rand ** 2)) / np.size(rand)
    stop     = timeit.default_timer()
    
    data = data.append({'Number of points': points, 'Integral Value': integral, 'Real Error': error,
                        'Statistical Error': stat_error, 'Runtime': (stop - start)}, ignore_index = True)

# data.to_excel("simple.xlsx")
data

```

### Important sampling

```python
real_integral = 0.882081
a = 0
b = 2
points_mat = np.arange(1e5, 1e6+1e5, 1e5)
data = pd.DataFrame()
data = pd.DataFrame(columns = ['Number of points', 'Integral Value', 'Real Error', 'Statistical Error', 'Runtime'])

for points in points_mat:

    start = timeit.default_timer()
    rand  = np.random.uniform(np.e ** -b, np.e ** -a, int(points))
    rand  = -np.log(rand)
    integral = np.mean(np.e ** (-rand ** 2 + rand)) * (1 - np.e ** -2)
    error = np.abs(integral - real_integral)
    stat_error = np.std(np.e ** (-rand ** 2 + rand)) / np.size(rand)
    stop     = timeit.default_timer()
    
    data = data.append({'Number of points': points, 'Integral Value': integral, 'Real Error': error,
                        'Statistical Error': stat_error, 'Runtime': (stop - start)}, ignore_index = True)

data.to_excel("important.xlsx")
data
```
