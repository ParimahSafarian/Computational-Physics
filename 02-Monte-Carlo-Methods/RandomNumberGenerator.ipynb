# Random Number Generator

Implementation and testing of pseudorandom number generators. Analyzes statistical properties and uniformity of generated sequences for Monte Carlo applications.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import random
import math
```

```python
n = 10000
rand = []

for i in range(n):
    rand.append(np.floor(random.random()*10))
    

plt.hist(rand, 10, density=True, facecolor='khaki')

plt.title("{} random number generated in range 0-9 histogram".format(n))
plt.ylabel("frequecy")
plt.xlabel("number")

plt.savefig("random.png")
```

```python
 def n_random_generator(n):
    rand = []

    for i in range(n):
        rand.append(np.floor(random.random()*10))
    
    return rand
```

```python
std = []
n   = 1500
for i in range(100, n):
    std.append(np.std(n_random_generator(i))/i)
    
plt.loglog(np.arange(100, n), std, c = 'forestgreen', label = r'$\frac{\sigma}{N} \propto N^{-0.5}$')

plt.title("mean standard deviation relation with N")
plt.ylabel(r'$\frac{\sigma}{N}$', size = 14)
plt.xlabel("$N$")
plt.legend()

plt.savefig("std.png")
```
