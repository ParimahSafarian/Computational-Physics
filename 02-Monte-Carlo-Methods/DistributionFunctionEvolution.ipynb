```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import random
import math
```

```python
m = 10
def n_guassian(n, mu, sigma):
    rand = []
    a = (mu / m - sigma * np.sqrt(3/m))
    b = (mu / m + sigma * np.sqrt(3/m))
    
    while np.size(rand) <= n:
        rand.append(np.sum(np.floor(np.random.uniform(a, b+1, m))))
 
    return rand
```

```python
%%time

color = ['orange', 'yellow', 'lawngreen', 'lightblue']

n      = 1000
mu_mat = [50, 100, 150, 200] 
sigma  = [20,  25,  50,  10]


for mu in mu_mat:
    Q = plt.figure()
    plt.hist(n_guassian(n, mu, sigma[mu_mat.index(mu)]), density = True, alpha = 0.8,
             facecolor = color[mu_mat.index(mu)], label = '$\mu ={}, \sigma = {}$'.format(mu, sigma[mu_mat.index(mu)]))

    plt.title("Distribution of {} numbers generated by my guassian number generator".format(n))
    plt.ylabel("frequecy")
    plt.xlabel("number")
    plt.legend()

    plt.savefig("{}.png".format(mu_mat.index(mu)))
```
