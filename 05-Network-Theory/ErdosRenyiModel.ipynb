# Erdős–Rényi Random Graph Model

Generation and analysis of random graphs using the Erdős–Rényi model. Studies connectivity, clustering, and emergence of giant components as a function of edge probability.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import random
import math
from networkx import nx
```

### Graph drawing

```python
n = 500    # nodes
k_mat = [0.8, 1, 8] # average edge number
g = []

for k in k_mat:
    m = n * k  # edges
    g.append(nx.gnm_random_graph(n, m))
    Q = nx.draw(g[k_mat.index(k)])
    plt.title(f"Erdos Renyi graph for {n} nodes with {k} average number of edge")
    plt.savefig(f"graph{k}.png")
    plt.show()
```

### Degree distribution

```python
%%time

for i in range(np.size(g)):
    degree = []
    for node in nx.nodes(g[i]):
        degree.append(nx.degree(g[i])[node])
    
    plt.hist(degree, 10, label = f'<k> = {k_mat[i]}')

    plt.title("Degree distribution for different values of <k>")
    plt.xlabel("degree")
    plt.ylabel("frequency")
    plt.savefig("degree.png")

    plt.legend()
```

### Clustering distribution 

```python
%%time

for i in range(np.size(g)):
    clustering = []
    for node in nx.nodes(g[i]):
        clustering.append(nx.clustering(g[i])[node])
    
    plt.hist(clustering, 20, label = f'<k> = {k_mat[i]}')

    plt.title("Clustering distribution for different values of <k>")
    plt.xlabel("clustering")
    plt.ylabel("frequency")
    plt.savefig("clustering.png")

    plt.legend()
```
