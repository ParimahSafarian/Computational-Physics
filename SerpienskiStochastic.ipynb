```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
#defining some functions to scale, rotait and transit
######

def Serpienski1(x , y):
    x1 = np.array(1/2) * x 
    y1 = np.array(1/2) * y  
    return x1 , y1

def Serpienski2(x , y):
    x2 = np.array(1/2) * x + np.array(1/2)
    y2 = np.array(1/2) * y
    return x2 , y2

def Serpienski3(x , y):
    x3 = np.array(1/2) * x + np.array(1/4)
    y3 = np.array(1/2) * y + np.array(np.sqrt(3)/4)
    return x3 , y3
```

```python
#plot    
Q = plt.figure(figsize=(8, 8))
plt.axis('equal')

# our initial black triangle
x = np.array([0 , 0.5 , 1])
y = np.array([0 , np.sin(np.pi/3) , 0])
plt.fill(x , y , 'black' )
```

```python
#plt.fill(x, y, 'black')
for j in range(10000):
    x = np.array([0 , 0.5 , 1])
    y = np.array([0 , np.sin(np.pi/3) , 0])
    for i in range(7):
        rand = random.random() * 3
        if rand < 1:
            x, y = Serpienski1(x , y)
        elif rand < 2:
            x, y = Serpienski2(x , y)
        elif rand <= 3:
            x, y = Serpienski3(x , y)
    plt.fill(x, y, 'black')
    #print(x, y)
    
plt.title("Stochastic Serpienski Triangle")
plt.savefig('Stochastic.png')
```
