```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#defining some functions to scale, rotait and transit


def Serpienski1( x , y):
    x1 = np.array(1/2) * x 
    y1 = np.array(1/2) * y  
    return x1 , y1

def Serpienski2(x , y):
    x2 = np.array(1/2) * x + np.array(1/2)
    y2 = np.array(1/2) * y
    return x2 , y2

def Serpienski3(x , y):
    x3 = np.array(1/2) * x + np.array(1/4)
    y3 = np.array(1/2) * y + np.array(np.sqrt(3)/4)
    return x3 , y3
```

```python
#plot    
Q = plt.figure(figsize=(8, 8))
plt.axis('equal')

# our initial black triangle
x = np.array([0 , 0.5 , 1])
y = np.array([0 , np.sin(np.pi/3) , 0])
plt.fill(x , y , 'black' )

plt.savefig('Step 0')
```

```python
#plot    
Q = plt.figure(figsize=(8, 8))
plt.axis('equal')

# our initial black triangle
x = np.array([0 , 0.5 , 1])
y = np.array([0 , np.sin(np.pi/3) , 0])
plt.fill(x , y , 'black' )


x_new = np.array([0.25 , 0.75 , 0.5])
y_new = np.array([np.sin(np.pi/3)/2  , np.sin(np.pi/3)/2 , 0])
plt.fill(x_new , y_new , 'white' )

plt.savefig('Step 1')
```

```python
#Level of Serpienski
run_num = 6
for l in range(run_num):
    Q = plt.figure(figsize=(8, 8))
    plt.axis('equal')

    # our initial black triangle
    x = np.array([0 , 0.5 , 1])
    y = np.array([0 , np.sin(np.pi/3) , 0])
    plt.fill(x , y , 'black' )
    #plt.show()
    #plt.savefig('Step 0')

    x_new = np.array([0.25 , 0.75 , 0.5])
    y_new = np.array([np.sin(np.pi/3)/2  , np.sin(np.pi/3)/2 , 0])
    plt.fill(x_new , y_new , 'white' )
    #plt.savefig('Step 1')
    
    for i in range(l):
        x = x_new
        y = y_new
        x1 , y1 = Serpienski1(x , y)
        x2 , y2 = Serpienski2(x , y)
        x3 , y3 = Serpienski3(x , y)
        x_new = np.append(x_new , x1)
        x_new = np.append(x_new , x2)
        x_new = np.append(x_new , x3)
        y_new = np.append(y_new , y1)
        y_new = np.append(y_new , y2)
        y_new = np.append(y_new , y3)

    #
    n= len(x_new)
    n = n/3
    x_new = np.reshape(x_new , (int(n) , 3))
    y_new = np.reshape(y_new , (int(n) , 3))

    for i in range(0 , 2**(l*2) , 1):
        plt.fill(x_new[i] , y_new[i] , 'white')
    plt.savefig('Step {}'.format(l+1))
    plt.show()
```
