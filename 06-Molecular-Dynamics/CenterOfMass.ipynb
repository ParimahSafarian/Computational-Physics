# Center of Mass Analysis

Analysis of center of mass motion and conservation laws in many-body systems. Studies translational and rotational dynamics, momentum conservation, and system stability.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import random
import math
```

```python
real_integral = 1/15
a = -1
b =  1
points = 1e7

rand       = np.random.uniform(a, b, int(points))
u_integral = np.mean((1-rand ** 2) * rand * (3+rand)/4) * (b-a)
u_error      = np.abs(u_integral - real_integral)
u_stat_error = np.std((1-rand ** 2) * rand * (3+rand)/4) / np.size(rand)

rand       = np.random.uniform(a, b, int(points))
d_integral = np.mean((1-rand ** 2) * (3+rand)/4) * (b-a)
d_error      = np.abs(d_integral - 1)
d_stat_error = np.std((1-rand ** 2) * (3+rand)/4) / np.size(rand)


y_cm = u_integral / d_integral
 
print(f'calculated y is {np.round(y_cm, 8)} and real value of it is {np.round(real_integral, 8)}')
```
