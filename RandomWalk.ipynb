```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import random
import math
```

```python
time  = np.arange(1000)
probs = [0.1, 0.3, 0.5, 0.7, 0.9]
ave   = 200
x     = np.zeros((ave, np.size(time)))
color = ['orange', 'deepskyblue', 'lawngreen', 'gold', 'dodgerblue']
x_m   = np.zeros((np.size(probs), np.size(time)))
```

```python
for p in range(np.size(probs)):
    for step in range(np.size(time)-1):
        for m in range(ave):
            if random.random() <= probs[p]:
                x[m, step+1] = x[m, step] + 1
            else:
                x[m, step+1] = x[m, step] - 1
                
        x_m[p, step+1] = np.mean(x[:, step+1])
        
    plt.scatter(time, x_m[p, :], s = 0.5, c = color[p], label = '(p-q) = {}'.format(np.round((2 * probs[p]-1), 2)))
    
plt.title("$<x_{(t)}>$ in 1D random walk\n for different probabilities")
plt.ylabel("$<x>$")
plt.xlabel("$time$")

plt.legend()
      
plt.savefig("xmean.png")
```

```python
s_x   = np.zeros((ave, np.size(time)))
s_m   = np.zeros((np.size(probs), np.size(time)))

for p in range(np.size(probs)):
    for step in range(np.size(time)-1):
        for m in range(ave):
            if random.random() <= probs[p]:
                s_x[m, step+1] = s_x[m, step] + 1
            else:
                s_x[m, step+1] = s_x[m, step] - 1
                
        s_m[p, step+1] = np.std(s_x[:, step+1]) ** 2
        
    plt.scatter(time, s_m[p, :], s = 0.5, c = color[p], label = 'pq = {}'.format(np.round(probs[p]*(1-probs[p]), 2)))
    plt.plot(time, 4 * time * probs[p]*(1-probs[p]), c = 'black', alpha = 0.5)
    
plt.title("$\sigma^2_{(t)}$ in 1D random walk\n for different probabilities")
plt.ylabel("$\sigma^2_{(t)}$")
plt.xlabel("$time$")

plt.legend()

plt.savefig("var.png")
```
