```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import random
import math
```

```python
def random_walk_lifetime(x_0, trap):
    p = 0.5
    ave = 500
    lifetime = np.zeros(ave)
    for m in range(ave):
        x = x_0
        while x <= trap and x >= -trap:
            lifetime[m] += 1
            if random.random() < p:
                x += 1
            else:
                x -= 1
                
    return np.mean(lifetime[:])
```

```python
print(random_walk_lifetime(20, 20))
```

```python
%%time

color = ['lawngreen', 'deepskyblue', 'orangered', 'gold', 'dodgerblue']

trap_mat  = [10, 15, 20]


for trap in trap_mat:
    initial_x = np.arange(-trap, trap+1, 1)
    mean_lifetime = np.zeros(np.size(initial_x))
    
    for k in range(np.size(initial_x)):
        mean_lifetime[k] = random_walk_lifetime(initial_x[k], trap)
    
    t = trap_mat.index(trap)
    plt.plot(initial_x, mean_lifetime, '.', c = color[t], label = '{}<x<{}'.format(-trap, trap))
    plt.plot(initial_x, mean_lifetime, c = color[t], dashes=[10,5] , linewidth = 0.3)


plt.title('Mean lifetime for 1D random walk with different boundaries versus initial position\n averaged on 500 runs')
plt.ylabel('Lifetime')
plt.xlabel('Initial position')
plt.legend()
plt.savefig('lifetime.png')
```
